```python
import pandas as pd
import shapefile
from json import dumps
import os
import json
import numpy as np
```

```python
pd.options.display.max_columns=50
pd.options.display.max_rows=50
```

```python
#r1filepath and r2 filepath should point directly to geojson files of all subwatersheds in HUC Region 1 and 2
#The underlying shapefiles for these subwatersheds can be found at https://www.usgs.gov/national-hydrography/access-national-hydrography-products

r1path = 'filepath1'
r2path = 'filepath2'
destintion = 'filepath3'
```

# Reading in HUCs from region 1

```python
assert os.path.exists(r1path), "Input file does not exist."
```

```python
f = open(r1path)
r1 = json.load(f)
```

# Region 1 Selection

```python
df=pd.json_normalize(r1['features'])
```

```python
#Stating with the Connecticut Coastal Subregion
LIS=df.loc[df["properties.huc12"].str.startswith("011000")]
```

```python
#Getting number of subwatersheds
print(len(LIS))
```

```python
#Looking for Pawcatuck river 
#Seems like it is actually in above subregion despite HUC of 01090005
df.loc[df["properties.huc12"].str.startswith("010900")]
```

```python
#Adding Lower Connecticut, Farmington, and Pawcatuck Subbasins
r1_additions=df.loc[(df["properties.huc12"].str.startswith("01080205")) | (df["properties.huc12"].str.startswith("01080207")) | (df["properties.huc12"].str.startswith("01090005"))]
r1_additions
```

```python
print(len(r1_additions))
```

```python
#Concatenating
LIS=pd.concat([LIS, r1_additions], axis=0)
print(len(LIS))
```

# Reading in HUCs from Region 2

```python
assert os.path.exists(r2path), "Input file does not exist."
```

```python
f= open(r2path)
r2 = json.load(f)
```

# Region 2 Selection

```python
df2=pd.json_normalize(r2['features'])
```

```python
#Looking for LIS itself
df2.loc[df2["properties.huc12"].str.startswith("02030203")]
```

```python
#Adding in Southern LIS, Northern LIS, and Bronx Subbasins + LIS itself
r2_additions=df2.loc[(df2["properties.huc12"].str.startswith("02030202")) | (df2["properties.huc12"].str.startswith("02030201")) | (df2["properties.huc12"].str.startswith("02030102")) | (df2["properties.huc12"].str.startswith("02030203"))]
r2_additions
```

```python
print(len(r2_additions))
```

```python
#Concatenating
LIS=pd.concat([LIS, r2_additions], axis=0)
```

```python
print(len(LIS))
```

```python
#Checking for duplicates
LIS.loc[LIS.duplicated(subset="properties.huc12")]
```

```python
#Selecting proper columns for table output
output=LIS.iloc[:, 6:-1].reset_index(drop=True)
output
```

# Output

```python
#The output will be a csv table with subwatershed areas and names included as columns
#The geometry column with the actual shape has been removed for simplicity
output.to_csv(destintion)
```
